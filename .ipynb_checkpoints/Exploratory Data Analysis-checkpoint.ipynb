# Titanic Data Set - Exploratory Data Analysis


The aim of this exercise is to carry out an exploratory data analysis which allows to find some patrons and correlations between the data set. This information would then be used to generate a Machine Learning that'll predict if a new entry would survive the Titanic sinking.

The aim of this exercise is to practice Data Science skill and upload predicted values to Kaggle.com

### data set columns:

 - Survival: Whether the passenger survived (0 or 1) .
 - Pclass: Ticket class (1, 2, 3).
 - Sex: Male or Female.
 - Age: Age in years.
 - SibSp: Number of siblings/spouses aboard the titanic.
 - Parch: Number of parens/children aboard the titanic.
 - Ticket: Ticket Number.
 - Fare: Passenger Fare.
 - Cabin: Cabin number.
 - Embarked: Port of Embarkation (C = Cherbourg, Q = Queenstown, S = Southampton)

source: https://www.kaggle.com/c/titanic/data


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats 

plt.rcParams['figure.figsize'] = (12,12)

data_set = pd.read_csv("./train.csv")
n_entries = data_set.shape[0]
```

## Missing Values

Check the porcentage of null values for each column on the data set.

```python
def checkNullValues():
    for column in data_set:
        print("{} | {} | {}".format(
            data_set[column].name,
            len(data_set[data_set[column].isnull()]) / (1.0*n_entries),
            data_set[column].dtype
        ))
checkNullValues()
```

### Outliers

For the matter of this analysis, an outlier will be a value with Z score greater than 3.

Z score formula:

$$z(x)= \frac{x-\mu}{\sigma}$$

```python
def outliers_col():
    for column in data_set:
        if data_set[column].dtype != np.object:
            n_outliers = len(data_set[np.abs(stats.zscore(data_set[column])) > 3])    
            print("{} | {} | {}".format(
                data_set[column].name,
                n_outliers,
                data_set[column].dtype
        ))
outliers_col()
```
