# Titanic Data Set - Exploratory Data Analysis


The aim of this exercise is to carry out an exploratory data analysis which allows to find some patterns and correlations between the data set. This information would then be used to generate a Machine Learning that'll predict if a new entry would survive the Titanic sinking.

The aim of this exercise is to practice Data Science skill and upload predicted values to Kaggle.com

### data set columns:

 - Survival: Whether the passenger survived (0 or 1) .
 - Pclass: Ticket class (1, 2, 3).
 - Sex: Male or Female.
 - Age: Age in years.
 - SibSp: Number of siblings/spouses aboard the titanic.
 - Parch: Number of parens/children aboard the titanic.
 - Ticket: Ticket Number.
 - Fare: Passenger Fare.
 - Cabin: Cabin number.
 - Embarked: Port of Embarkation (C = Cherbourg, Q = Queenstown, S = Southampton)

source: https://www.kaggle.com/c/titanic/data


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats 
import seaborn as sns

df = pd.read_csv("./train.csv")
n_entries = df.shape[0]
```

```python
n_entries
```

```python
df.dtypes
```

## Missing Values

Check the porcentage of null values for each column on the data set.

```python
def checkNullValues():
    for column in df:
        print("{} | {} | {}".format(
            df[column].name,
            len(df[df[column].isnull()]) / (1.0*n_entries),
            df[column].dtype
        ))
checkNullValues()
```

## Attribute Distributions

```python
categorical_variables = ['Pclass', 'Sex', 'Embarked']
numerical_variables = ['Age', 'SibSp', 'Parch', 'Fare']

background = '#98deed'
palette = ["#800000","#8000ff","#6aac90","#5833ff","#da8829"]

fig = plt.figure(figsize=(16, 20))

fig.patch.set_facecolor(background)
gs = fig.add_gridspec(4,2)

ax0 = fig.add_subplot(gs[0, 0])
ax1 = fig.add_subplot(gs[0, 1])
ax2 = fig.add_subplot(gs[1, 0])
ax3 = fig.add_subplot(gs[1, 1])

# Plot title
ax0.spines['bottom'].set_visible(False)
ax0.spines['top'].set_visible(False)
ax0.spines['right'].set_visible(False)
ax0.spines['left'].set_visible(False)
ax0.tick_params(left=False, bottom=False)
ax0.set_xticklabels([])
ax0.set_yticklabels([])

ax0.text(0.5,0.5,
         'Count plot for various\n categorical features\n_________________',
         horizontalalignment='center',
         verticalalignment='center',
         fontsize=18, fontweight='bold',
         fontfamily='serif',
         color="#000000")
ax0.set_facecolor(background)

# Pclass plot
sns.countplot(ax=ax1, x='Pclass', data=df, palette=palette, fill=True)
ax1.grid(color='#000000', linestyle=':', axis='y', zorder=0,  dashes=(1,5))

# Sex plot
sns.countplot(ax=ax2, x='Sex', data=df, palette=palette, fill=True)
ax2.grid(color='#000000', linestyle=':', axis='y', zorder=0,  dashes=(1,5))

# Embarked plot
sns.countplot(ax=ax3, x='Embarked', data=df, palette=palette, fill=True)
ax3.grid(color='#000000', linestyle=':', axis='y', zorder=0,  dashes=(1,5))
```

```python
fig = plt.figure(figsize=(18, 35))

gs = fig.add_gridspec(8,2)
fig.set_facecolor(background)

ax0 = fig.add_subplot(gs[0,0])
ax1 = fig.add_subplot(gs[0,1])
ax2 = fig.add_subplot(gs[1,0])
ax3 = fig.add_subplot(gs[1,1])
ax4 = fig.add_subplot(gs[2,0])
ax5 = fig.add_subplot(gs[2,1])
ax6 = fig.add_subplot(gs[3,0])
ax7 = fig.add_subplot(gs[3,1])

# Age plots
sns.kdeplot(ax=ax0, data=df, x='Age', fill=True, palette=palette)
sns.boxplot(ax=ax1, data=df, x='Age')

# SibSp plots
sns.kdeplot(ax=ax2, data=df, x='SibSp', fill=True, palette=palette)
sns.boxplot(ax=ax3, data=df, x='SibSp')

# Parch plots
sns.kdeplot(ax=ax4, data=df, x='Parch', fill=True, palette=palette)
sns.boxplot(ax=ax5, data=df, x='Parch')

# Fare plots
sns.kdeplot(ax=ax6, data=df, x='Fare', fill=True, palette=palette)
sns.boxplot(ax=ax7, data=df, x='Fare')
```

## Conclusions

 - Around 77% of Cabin values are missing, making it useless for training a model.
 - Around 20% of Age values are missing. However, this might still be useful in the context of a boat sinking, the age of an individual might make a change either because of maturity or safety boats priotiry. Apart from that this attribute seems to have a normal distribution, therefore missing values can be filled with the mean or median.
 - Around 65% of passengers were men and 35% women.
 - More than a half of passengers embarked from Southampton.
 - More than a half of passengers died.
 - Around a Half of the passengers were travelling in 3rd class.
 - The ticket fair seems to have some outliers as a considerable amount of people paid less than 100$.
 - Children were the most likely to survive.
 - It was noticed that even though there were more males in the boat, more females survived.
 - People with 1st and 2nd class tickets were most likely to survive, whereas 3rd class were the less likely with a chances of only 24%, it was also the class with more people.
 - Almost every numerical variable has outliers.
 - The only numerical variables with a normal distribution is age.
